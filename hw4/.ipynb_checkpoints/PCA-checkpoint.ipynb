```python
import numpy as np
import os
from scipy import misc
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

X = []
path = '~/Downloads/faceExpressionDatabase/'
for i in range(10):
    letter = chr(ord('A') + i )
    for j in range(10):
        image = misc.imread(os.path.join(path, letter + '0' + str(j) + '.bmp'), flatten = 0)
        X.append(image.reshape(64 * 64))
    #plt.imshow(image, cmap=plt.get_cmap('gray'))
    #plt.show()


X = np.array(X)
X_mean = X.mean(axis=0, keepdims=True)
#print(X_mean.shape)
plt.imshow(X_mean.reshape(64, 64), cmap=plt.get_cmap('gray'))
plt.show()
plt.clf()

u, s, v = np.linalg.svd(X - X_mean)

gs = gridspec.GridSpec(3, 3, top=1., bottom=0., right=1., left=0., hspace=0.1, wspace=0.)
i = 0
for g in gs:
    ax = plt.subplot(g)
    ax.imshow(v[i].reshape(64, 64), cmap=plt.get_cmap('gray'))
    ax.set_xticks([])
    ax.set_yticks([])
    i = i + 1
plt.show()
```

```python
gs = gridspec.GridSpec(10, 10, top=1., bottom=0., right=1., left=0., hspace=0.1, wspace=0.)
plt.figure(figsize=(10, 10))
i = 0
for g in gs:
    ax = plt.subplot(g)
    ax.imshow(X[i].reshape(64, 64), cmap=plt.get_cmap('gray'))
    ax.set_xticks([])
    ax.set_yticks([])
    i = i + 1
plt.show()
```

```python
mu = X.mean(axis=0)

Z = []
for i in range(100):   
    recotX = np.zeros(64 * 64)
    for j in range(5):
        recotX = recotX + np.dot((X[i] - mu), v[j]) * v[j]   
    recotX = recotX + mu
    Z.append(recotX)

gs = gridspec.GridSpec(10, 10, top=1., bottom=0., right=1., left=0., hspace=0.1, wspace=0.)
plt.figure(figsize=(10, 10))
i = 0
for g in gs:
    ax = plt.subplot(g)
    ax.imshow(Z[i].reshape(64, 64), cmap=plt.get_cmap('gray'))
    ax.set_xticks([])
    ax.set_yticks([])
    i = i + 1
plt.show()
```

```python
mu = X.mean(axis=0)
for k in range(1, 100):
    mse = 0
    for i in range(100):
        recotX = np.zeros(64 * 64)
        for j in range(k):
            recotX = recotX + np.dot((X[i] - mu), v[j]) * v[j]   
        recotX = recotX + mu
        mse = mse + np.mean((recotX - X[i])**2)
    rmse = np.sqrt(mse / 100)
    print(rmse)
    if(rmse < 1):
        print(k)
        break
```
